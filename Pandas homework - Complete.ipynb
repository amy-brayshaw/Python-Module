```python
plt.figure(figsize=(10,10))
plt.pie(
    game_sales,
    labels=game_sales.index,
    autopct=lambda p:'{:.0f}'.format(p * game_sales.sum() / 100),
    startangle=90,
    textprops={'fontsize': 10}
)

plt.title('Total Sales by Game: Top 10', fontsize=16)
plt.tight_layout()

plt.show()
```

```python
game_sales = n64bestsellers_df.groupby('Game')['Sales'].sum().reset_index()
game_sales = game_sales.sort_values(by='Sales', ascending=False)[0:10]
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=game_sales, x='Game', y='Sales', palette='viridis', hue='Game', legend=False)

plt.title('Total Sales: Top 10 Best Sellers', fontsize=16)
plt.xlabel('Game', fontsize=12)
plt.ylabel('Total Sales (Millions)', fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
publisher_sales = n64bestsellers_df.groupby('Publisher(s)')['Sales'].sum().reset_index()
publisher_sales = publisher_sales.sort_values(by='Sales', ascending=False)
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=publisher_sales, x='Publisher(s)', y='Sales', palette='viridis', hue='Publisher(s)', legend=False)

plt.title('Total Sales by Publisher', fontsize=16)
plt.xlabel('Publisher(s)', fontsize=12)
plt.ylabel('Total Sales (Millions)', fontsize=12)
plt.xticks(rotation=45, ha='right')

plt.tight_layout()
plt.show()
```
